# Data import

```python
h3_histogram = pd.read_csv('report/H3_histogram.csv')
h4a_histogram = pd.read_csv('report/H4A_histogram.csv')
h19_histogram = pd.read_csv('report/H19_histogram.csv')

h3_raw = pd.read_csv('report/H3.csv')
h4a_raw = pd.read_csv('report/H4A.csv')
h19_raw = pd.read_csv('report/H19.csv')

h3 = h3_raw[300:3900]
h4a = h4a_raw[300:3900]
h19 = h19_raw[300:3900]

h3_pores_raw = pd.read_csv('report/H3_pores.csv')
h4a_pores_raw = pd.read_csv('report/H4A_pores.csv')
h19_pores_raw = pd.read_csv('report/H19_pores.csv')

h3_pores = h3_pores_raw[300:3900]
h4a_pores = h4a_pores_raw[300:3900]
h19_pores = h19_pores_raw[300:3900]

h3_id = np.array(h3[' '])
h4a_id = np.array(h4a[' '])
h19_id = np.array(h19[' '])

h3_length = 20
h4a_length = 20
h19_length = 20

h3_por = np.array(h3['%Area'])
h3_por_max = np.max(h3_por)
h3_por_max_id = h3_id[np.argmax(h3_por)]
h3_por_min = np.min(h3_por)
h3_por_min_id = h3_id[np.argmin(h3_por)]

h3_cov = np.array(h3['StdDev']) / np.array(h3['Mean'])
h3_cov_max = np.max(h3_cov)
h3_cov_max_id = h3_id[np.argmax(h3_cov)]
h3_cov_min = np.min(h3_cov)
h3_cov_min_id = h3_id[np.argmin(h3_cov)]

h4a_por = np.array(h4a['%Area'])
h4a_por_max = np.max(h4a_por)
h4a_por_max_id = h4a_id[np.argmax(h4a_por)]
h4a_por_min = np.min(h4a_por)
h4a_por_min_id = h4a_id[np.argmin(h4a_por)]

h4a_cov = np.array(h4a['StdDev']) / np.array(h4a['Mean'])
h4a_cov_max = np.max(h4a_cov)
h4a_cov_max_id = h4a_id[np.argmax(h4a_cov)]
h4a_cov_min = np.min(h4a_cov)
h4a_cov_min_id = h4a_id[np.argmin(h4a_cov)]

h19_por = np.array(h19['%Area'])
h19_por_max = np.max(h19_por)
h19_por_max_id = h19_id[np.argmax(h19_por)]
h19_por_min = np.min(h19_por)
h19_por_min_id = h19_id[np.argmin(h19_por)]

h19_cov = np.array(h19['StdDev']) / np.array(h19['Mean'])
h19_cov_max = np.max(h19_cov)
h19_cov_max_id = h19_id[np.argmax(h19_cov)]
h19_cov_min = np.min(h19_cov)
h19_cov_min_id = h19_id[np.argmin(h19_cov)]

h3_count = np.array(h3_pores['Count'])
h3_count_max = np.max(h3_count)
h3_count_max_id = h3_id[np.argmax(h3_count)]
h3_count_min = np.min(h3_count)
h3_count_min_id = h3_id[np.argmin(h3_count)]

h3_feret_diameter = np.array(h3_pores['Feret'])
h3_feret_diameter_max = np.max(h3_feret_diameter)
h3_feret_diameter_max_id = h3_id[np.argmax(h3_feret_diameter)]
h3_feret_diameter_min = np.min(h3_feret_diameter)
h3_feret_diameter_min_id = h3_id[np.argmin(h3_feret_diameter)]

h3_eq_pore_diameter = 3 / 2 * ((np.array(h3_pores['Feret']) + np.array(h3_pores['MinFeret'])) / 2)
h3_eq_pore_diameter_max = np.max(h3_eq_pore_diameter)
h3_eq_pore_diameter_max_id = h3_id[np.argmax(h3_eq_pore_diameter)]
h3_eq_pore_diameter_min = np.min(h3_eq_pore_diameter)
h3_eq_pore_diameter_min_id = h3_id[np.argmin(h3_eq_pore_diameter)]

h3_ar = np.array(h3_pores['Minor']) / np.array(h3_pores['Major'])
h3_ar_max = np.max(h3_ar)
h3_ar_max_id = h3_id[np.argmax(h3_ar)]
h3_ar_min = np.min(h3_ar)
h3_ar_min_id = h3_id[np.argmin(h3_ar)]

h3_circularity = np.array(h3_pores['Circ.'])
h3_circularity_max = np.max(h3_circularity)
h3_circularity_max_id = h3_id[np.argmax(h3_circularity)]
h3_circularity_min = np.min(h3_circularity)
h3_circularity_min_id = h3_id[np.argmin(h3_circularity)]

h4a_count = np.array(h4a_pores['Count'])
h4a_count_max = np.max(h4a_count)
h4a_count_max_id = h4a_id[np.argmax(h4a_count)]
h4a_count_min = np.min(h4a_count)
h4a_count_min_id = h4a_id[np.argmin(h4a_count)]

h4a_feret_diameter = np.array(h4a_pores['Feret'])
h4a_feret_diameter_max = np.max(h4a_feret_diameter)
h4a_feret_diameter_max_id = h4a_id[np.argmax(h4a_feret_diameter)]
h4a_feret_diameter_min = np.min(h4a_feret_diameter)
h4a_feret_diameter_min_id = h4a_id[np.argmin(h4a_feret_diameter)]

h4a_eq_pore_diameter = 3 / 2 * ((np.array(h4a_pores['Feret']) + np.array(h4a_pores['MinFeret'])) / 2)
h4a_eq_pore_diameter_max = np.max(h4a_eq_pore_diameter)
h4a_eq_pore_diameter_max_id = h4a_id[np.argmax(h4a_eq_pore_diameter)]
h4a_eq_pore_diameter_min = np.min(h4a_eq_pore_diameter)
h4a_eq_pore_diameter_min_id = h4a_id[np.argmin(h4a_eq_pore_diameter)]

h4a_ar = np.array(h4a_pores['Minor']) / np.array(h4a_pores['Major'])
h4a_ar_max = np.max(h4a_ar)
h4a_ar_max_id = h4a_id[np.argmax(h4a_ar)]
h4a_ar_min = np.min(h4a_ar)
h4a_ar_min_id = h4a_id[np.argmin(h4a_ar)]

h4a_circularity = np.array(h4a_pores['Circ.'])
h4a_circularity_max = np.max(h4a_circularity)
h4a_circularity_max_id = h4a_id[np.argmax(h4a_circularity)]
h4a_circularity_min = np.min(h4a_circularity)
h4a_circularity_min_id = h4a_id[np.argmin(h4a_circularity)]

h19_count = np.array(h19_pores['Count'])
h19_count_max = np.max(h19_count)
h19_count_max_id = h19_id[np.argmax(h19_count)]
h19_count_min = np.min(h19_count)
h19_count_min_id = h19_id[np.argmin(h19_count)]

h19_feret_diameter = np.array(h19_pores['Feret'])
h19_feret_diameter_max = np.max(h19_feret_diameter)
h19_feret_diameter_max_id = h19_id[np.argmax(h19_feret_diameter)]
h19_feret_diameter_min = np.min(h19_feret_diameter)
h19_feret_diameter_min_id = h19_id[np.argmin(h19_feret_diameter)]

h19_eq_pore_diameter = 3 / 2 * ((np.array(h19_pores['Feret']) + np.array(h19_pores['MinFeret'])) / 2)
h19_eq_pore_diameter_max = np.max(h19_eq_pore_diameter)
h19_eq_pore_diameter_max_id = h19_id[np.argmax(h19_eq_pore_diameter)]
h19_eq_pore_diameter_min = np.min(h19_eq_pore_diameter)
h19_eq_pore_diameter_min_id = h19_id[np.argmin(h19_eq_pore_diameter)]

h19_ar = np.array(h19_pores['Minor']) / np.array(h19_pores['Major'])
h19_ar_max = np.max(h19_ar)
h19_ar_max_id = h19_id[np.argmax(h19_ar)]
h19_ar_min = np.min(h19_ar)
h19_ar_min_id = h19_id[np.argmin(h19_ar)]

h19_circularity = np.array(h19_pores['Circ.'])
h19_circularity_max = np.max(h19_circularity)
h19_circularity_max_id = h19_id[np.argmax(h19_circularity)]
h19_circularity_min = np.min(h19_circularity)
h19_circularity_min_id = h19_id[np.argmin(h19_circularity)]
```
